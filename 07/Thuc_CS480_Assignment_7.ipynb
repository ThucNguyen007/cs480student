```python
# MPR is a method of reconstructing images on all three axial, coronal and sagittal planes. 
# Commonly known as Multiplanar Reconstruction
```

```python
# Question 2): What is the difference between MIP, MinIP, AvgIP?
# 
```

```python
# Commonly known as Maximum Intensity Projection.
# MIP is a mathematical method by which the reinforcement voxels are extracted from two-dimensional or three-dimensional datasets. 
# Voxels are selected from a set of data obtained from iodine at various angles. Then, projected as two-dimensional images.
# This method is often used in the study of blood vessels with contrast enhancement.
```

```python
# Minimum Intensity Projection - MinIP: The MinIP process is very similar to the MIP, but instead of displaying voxels with the highest density, it shows the voxels of the lowest density. 
# The main application of MinIP is to watch gas in the airways, parenchyma necrosis.
```
